# Data Analytics Interview Practice

Often for data analytics professionals, an analysis of a dataset is commonly required as part of an interview process. The following exercise below is an example of one where it involves joining reading and joining datasets together. Steps in carrying out this analysis include exploring the dataset and ensuring that data is tidy. Once this has been achieved we then start to frame questions and hypothesis to test with this data. It is common to jump straight into a dataset and explore it however my belief that for a well rounded approach, both are taken into the equation being carrying out hypothesis testing and carrying out EDA. 

The data below refers to the individual grocery shipping ticket data containing items purchased by a customer during a specific transaction. 

So lets begin our journey. 

### 1. Import the neccessary modules

```python
# Importing key modules for use
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  
import numpy as np
%matplotlib inline
from empiricaldist import Pmf
from empiricaldist import Cdf
from scipy.stats import norm
```

### 2. Obtain the data from the csv files

```python
# read the data sources and save into dataframe
categories = pd.read_csv('categories.csv')
day_of_week = pd.read_csv('day_of_week.csv')
departments = pd.read_csv('departments.csv')
products = pd.read_csv('products.csv')
transactions = pd.read_csv('transactions.csv')
```

### 3. Feature Engineering

```python
# Can also use type(), dtypes, value_counts(), describe()

transactions.info()
print('-'*100)
categories.info()
print('-'*100)
day_of_week.info()
print('-'*100)
departments.info()
print('-'*100)
products.info() 
print('-'*100)
```

```python
# Merge dataframes together
# The source of the dataframes are rerieved from a relational database so they are left joined onto their primary key
t_1 = pd.merge(transactions,day_of_week, how='left', on='day_of_week_no')
t_2 = pd.merge(t_1,products,how='left',on = 'product_id')
t_3 = pd.merge(t_2,categories, how='left',on = 'category_id')
t_4 = pd.merge(t_3,departments, how='left', on = 'department_id')

trans = t_4
```

```python
# Summary Statistics

# Testing for null values
trans.isnull().sum()
# Results show that dataframe is clean with no null values. 
```

```python
# Checking the columns of the dataframe
trans.columns
```

#### a) Remapping

From below, we can see that the day of week name starts from 0 and corresponds to Saturday. We will remap it to begin from monday as a category 1. 

```python
# Checking the current mapping
trans[['day_of_week_name','day_of_week_no']].drop_duplicates()
```

```python
# create a dictionary for the remapping
dic = {'Mon':1, 'Tue':2, 'Wed':3, 'Thur':4, 'Fri':5, 'Sat':6, 'Sun':7}

# create a copy of column day_of week_name and remap it based on dictionary created
trans['remap'] = trans['day_of_week_name'].replace(dic)

#print(trans[['day_of_week_name','remap']].drop_duplicates()) # validate remapping

# Drop the depricated column
trans = trans.drop('day_of_week_no',axis =1)
trans = trans.rename(columns = {'remap':'day_of_week_no'}) # rename the column

trans[['day_of_week_name','day_of_week_no']].drop_duplicates()
```

```python
def conversion_rate(dataframe, column_names):
    # Total number of converted users
    column_conv = dataframe[dataframe['converted']==True].groupby(column_names)['user_id'].nunique()

    # Total number users
    column_total = dataframe.groupby(column_names)['user_id'].nunique()  
    
    # Conversion rate 
    conversion_rate = column_conv/column_total
    
    # Fill missing values with 0
    conversion_rate = conversion_rate.fillna(0)
    return conversion_rate
```

```python
def plotting_conv(dataframe):
    for column in dataframe:
        # Plot column by dataframe's index
        plt.plot(dataframe.index, dataframe[column])
        plt.title('Daily ' + str(column) + ' Pattern\n', 
                  size = 10)
        plt.ylabel('Count', size = 8)
        plt.xlabel('Date', size = 8)
        # Show plot
        plt.show()  
        plt.clf()
```
